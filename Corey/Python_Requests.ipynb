```python
import requests
from bs4 import BeautifulSoup

r = requests.get('https://imgs.xkcd.com/comics/python.png')

# write bytes
with open ('comic.png', 'wb') as f:
    f.write(r.content)
```

```python
# Check status....200 -> success, 300 -> redirecting
print(r.status_code)

# or.....less than 400
print(r.ok)
```
